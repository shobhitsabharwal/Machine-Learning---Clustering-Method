```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing dataset
dataset = pd.read_csv("Mall_customers.csv")
X = dataset.iloc[:, [3,4]].values
```

```python
#Using elblow method to determine optimal number of cluster
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmean= KMeans(n_clusters = i, max_iter=300, n_init = 10, init = 'k-means++', random_state=0)
    kmean.fit(X)
    wcss.append(kmean.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
#Optimum number of Cluster is 5
```

```python
#Applying k-Mean cluster with K=5
kmean= KMeans(n_clusters = 5, max_iter=300, n_init = 10, init = 'k-means++', random_state=0)
y_kmeans = kmean.fit_predict(X)
```

```python
#Ploting Clusters in scatter plot
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster1' )
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'Blue', label = 'Cluster2' )
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster3' )
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'magenta', label = 'Cluster4' )
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'cyan', label = 'Cluster5' )
plt.scatter(kmean.cluster_centers_[:,0], kmean.cluster_centers_[:,1], s = 200, c = 'Black', label = 'Centroid' )
plt.title("Clusters of clients")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```

```python
###  Insights   ###
#Cluster 1 High income , low spending -> carefull
#Cluster 2 Mid income , mid spending -> Standard
#Cluster 3 High income ,highspending -> Target
#Cluster 4 low income , high spending -> careless
#Cluster 5 low income , low spending -> Sensible
```

```python
#Ploting Clusters in scatter plot
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Carefull' )
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'Blue', label = 'Standard' )
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Target' )
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'magenta', label = 'Careless' )
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'cyan', label = 'Sensible' )
plt.scatter(kmean.cluster_centers_[:,0], kmean.cluster_centers_[:,1], s = 200, c = 'Black', label = 'Centroid' )
plt.title("Clusters of clients")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```
