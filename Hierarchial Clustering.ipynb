```python
"""
Created on Sat Mar 31 16:01:02 2018

@author: Shobhit Sabharwal
"""

#Clustering method
#Hierarchial Clustering

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:, [3,4]].values
```

```python
#using the dandrogram to find the optimal clusters numbers
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
```

```python
#insight
#Optimal number of clusters is 5
```

```python
#Fitting hierarchical cluster to dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
#visualising clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster1' )
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'Blue', label = 'Cluster2' )
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster3' )
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'magenta', label = 'Cluster4' )
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'cyan', label = 'Cluster5' )
plt.title("Clusters of clients")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```

```python
###  Insights   ###
#Cluster 1 High income , low spending -> carefull
#Cluster 2 Mid income , mid spending -> Standard
#Cluster 3 High income ,highspending -> Target
#Cluster 4 low income , high spending -> careless
#Cluster 5 low income , low spending -> Sensible
```

```python
#Ploting Clusters in scatter plot
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Carefull' )
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'Blue', label = 'Standard' )
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Target' )
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'magenta', label = 'Careless' )
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'cyan', label = 'Sensible' )
plt.title("Clusters of clients")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```
